```python
import os
import pandas as pd

directory_path = '/kaggle/input/march-machine-learning-mania-2024'
merged_df = pd.DataFrame()

for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory_path, filename)
        
        try:
            # Read the CSV file into a DataFrame with specified encoding
            df = pd.read_csv(file_path, encoding='latin-1')
            
            # Merge the current DataFrame with the main DataFrame
            merged_df = pd.concat([merged_df, df], ignore_index=True)
        except UnicodeDecodeError:
            print(f"Error reading file: {file_path}")

print(merged_df.head())
```

```python
import os
import pandas as pd
from tqdm import tqdm

directory_path = '/kaggle/input/march-machine-learning-mania-2024'
men_files = [
    'MSecondaryTourneyTeams.csv',
    'MSecondaryTourneyCompactResults.csv',
    'MNCAATourneySlots.csv',
    'MNCAATourneySeeds.csv',
    'MNCAATourneySeedRoundSlots.csv',
    'MNCAATourneyDetailedResults.csv',
    'MNCAATourneyCompactResults.csv'
]

women_files = [
    'WNCAATourneySlots.csv',
    'WNCAATourneySeeds.csv',
    'WNCAATourneyDetailedResults.csv',
    'WNCAATourneyCompactResults.csv',
]

# Initialize empty DataFrames for men's and women's tournament data
men_tourney_df = pd.DataFrame()
women_tourney_df = pd.DataFrame()

# Combine CSV files for men's tournament data
for men_file in tqdm(men_files, desc="Combining Men's Tournament Data"):
    file_path = os.path.join(directory_path, men_file)
    df = pd.read_csv(file_path)
    men_tourney_df = pd.concat([men_tourney_df, df], ignore_index=True)

# Combine CSV files for women's tournament data
for women_file in tqdm(women_files, desc="Combining Women's Tournament Data"):
    file_path = os.path.join(directory_path, women_file)
    df = pd.read_csv(file_path)
    women_tourney_df = pd.concat([women_tourney_df, df], ignore_index=True)

print("Combined Men's Tournament Data:")
men_tourney_df.head()
```
